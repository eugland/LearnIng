```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import Audio 
import scipy.io.wavfile 
Fs, y = scipy.io.wavfile.read('train_bird.wav') 
Audio(y, rate = Fs)
```

```python
fig, ((axis1, axis2), (axis3, axis4)) = plt.subplots(2, 2)

# 1. plot input Signal 
axis1.set_title("Input Signal")
axis1.set(xlabel = 'Time(Sample Index)', ylabel = 'Power')
axis1.plot(y)

# 2. Plot 
axis1.set_title("Power Spectrum of Input Signal")
axis1.set(xlabel = 'Frequency (Index)', ylabel = 'Power')

from scipy.fft import fft
freq_y = fft(y)
axis2.plot(freq_y)

```
